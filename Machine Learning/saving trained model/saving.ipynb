```python
import pandas as pd
import numpy as np
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
plt.xlabel('Area (sq ft)')
plt.ylabel('Price (USD)')
plt.title('Home Prices vs Area')
plt.scatter(df.area, df.price, color='red', marker='+', s=100)
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3500]])
```

```python
#to save the model we import pickle
import pickle
```

```python
with open('model_pickle', 'wb') as f:
    pickle.dump(reg, f)
```

```python
#to again use the model we load it and open it in read binary mode
with open('model_pickle', 'rb') as f:
    mp = pickle.load(f)
```

```python
mp.predict([[3500]])
#as we can see the prediction is same as before
```

```python
#if we want to save the model using joblib
import joblib

joblib.dump(reg, 'model_joblib')
#to load the model using joblib
mp2 = joblib.load('model_joblib')
mp2.predict([[3500]])
```
