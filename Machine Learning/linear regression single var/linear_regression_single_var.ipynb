Regression is the way to find the bestfit line with minumum errors

```python
import pandas as pd
import numpy as np
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
plt.xlabel('Area (sq ft)')
plt.ylabel('Price (USD)')
plt.title('Home Prices vs Area')
plt.scatter(df.area, df.price, color='red', marker='+', s=100)
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3500]])
```

```python
#y=mx+c is the equation of the line
# where m is the slope and c is the intercept
#reg.coef_  # This gives the slope (m)
#reg.intercept_  # This gives the intercept (c)
reg.coef_
```

```python
reg.intercept_
```

```python
# Now let's predict the price for a new area
d = pd.read_csv('areas.csv')
d.head()
```

```python
#this will predict the prices for the areas in 'areas.csv'
p = reg.predict(d)
d['prices'] = p
d.to_csv('predicted_prices.csv', index=False)
```

```python
# this will plot the regression line on the scatter plot
plt.scatter(df.area, df.price, color='red', marker='+', s=100)
plt.plot(df.area, reg.predict(df[['area']]), color='blue', linewidth=2)# this is the regression line
plt.grid()
plt.xlabel('Area (sq ft)')
plt.ylabel('Price (USD)')
plt.title('Home Prices vs Area with Regression Line')
plt.show()
```
