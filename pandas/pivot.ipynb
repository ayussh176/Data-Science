```python
import pandas as pd
df = pd.read_csv('weather_by_cities.csv')
df
```

```python
#this helped to reshape the data 
df.pivot(index='day',columns='city')
```

```python
#this give only windspeed with day as index and city as columns
df.pivot(index='day',columns='city',values='windspeed')
```

```python
df = pd.read_csv('weather2.csv')
df
#in this data frame i have 2 different values for same day 
```

```python
#in this the 2 different values are taken average
df.pivot_table(index='city',columns='date')
```

```python
#in this aggfunct help to decide either we have to do sum or average and adds a column and row for showing the 
#operational value for the row as well as column
df.pivot_table(index='city',columns='date',aggfunc='mean',margins=True)
```
