```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
#here
plt.plot(x,y,color='red',marker = "D",linewidth=1, linestyle='dashed')
```

```python
plt.plot(x,y,marker = "+",linestyle = '',markersize=10)
```

```python
plt.plot(x,y,color='red',marker = "D",linewidth=1, linestyle='dashed',alpha = 0.5)
#this is use to give transperency to the line it lies between 0-1
```

```python
plt.plot(x,y,color='red',marker = "X",markersize=10, linestyle='')
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel("days")
plt.ylabel("temperature")
plt.plot(days,max_t,marker='.')
plt.plot(days,avg_t,marker = '.')
plt.plot(days,min_t,marker='.')
plt.title("weather")
```

```python
plt.xlabel("days")
plt.ylabel("temperature")
plt.plot(days,max_t,marker='.',label = 'max')
plt.plot(days,avg_t,marker = '.',label='avg')
plt.plot(days,min_t,marker='.',label='min')
plt.title("weather")
plt.legend(loc ="lower left")
#loc is use to locate the location of legend box
#legend() help to specify the graph line
```
