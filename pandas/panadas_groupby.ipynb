```python
#we can concat two different data frames
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
cc = pd.concat([india_weather,us_weather],ignore_index=True)
cc
```

```python
cc = pd.concat([india_weather,us_weather],keys=['india','us'])
cc
```

```python
#we concat the two dataframes side by side
indiaweather = pd.DataFrame({
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
```

```python
usweather = pd.DataFrame({
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
```

```python
cc = pd.concat([indiaweather,usweather],axis=1,keys=['india','us'])
cc
```

```python
#we can give index to the dataframe
weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30]
    },index=[1,2,0])
weather
```

```python
humid = pd.DataFrame({
    "city": ["delhi","mumbai",],
    "temperature": [5,6]
    },index=[2,1])
humid
```

```python
z=pd.concat([weather,humid],axis=1)
z
```

```python
k = pd.Series(["humid","dry","cold"],name='event')
l = pd.concat([k,weather],axis=1)
l
```
