```python
import pandas as pd
import numpy as np
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```python
plt.scatter(df.year, df.per_capita_income, color='blue', marker='o', s=30)
plt.xlabel('Year')
plt.ylabel('Income in $CAD')
plt.title('Per Capita Income in Canada Over Years')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.per_capita_income)
```

```python
reg.predict([[2020]])
```
