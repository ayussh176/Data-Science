```python
import pandas as pd
df_stock = pd.read_excel("stock_data.xlsx","Sheet1")
df_stock
```

```python
def convert_people_cell(cell):
    if cell == "n.a.":
        return "sam walton"
    return cell
def convert_eps_cell(cell):
    if cell == "not available":
        return None
    return cell
def convert_revenue_cell(cell):
    if cell == -1:
        return 0
    return cell
def convert_price_cell(cell):
    if cell == "n.a.":
        return None
    return cell
df_stock = pd.read_excel("stock_data.xlsx","Sheet1",converters={'people':convert_people_cell,'eps':convert_eps_cell,'revenue':convert_revenue_cell,'price':convert_price_cell})
df_stock
```

```python
#this returns a ouput file as excel with name as new and shet name as stocks
#by default it gives the index
#start col and startrow leaves the enter spaces  
df_stock.to_excel('new.xlsx',sheet_name='stocks',index=False,startcol=1,startrow=2)
```

```python
df_weather = pd.read_csv('stock_data.csv')
df_weather
```

```python
with pd.ExcelWriter('stock_weather.xlsx') as writer:
    df_stock.to_excel(writer,sheet_name='stocks',index=False)
    df_weather.to_excel(writer,sheet_name='weather',index=False)
```
