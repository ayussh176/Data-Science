```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('insurance_data.csv')
df
```

```python
plt.scatter(df['age'], df['bought_insurance'])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test =train_test_split(df[['age']],df.bought_insurance, test_size=0.1, random_state=42)
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
#this tells the model to learn from the training data
```

```python
model.fit(X_train, y_train)
#this tells the model to predict the output for the test data
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
model.predict_proba(X_test)
#this gives the probability of each class
```
