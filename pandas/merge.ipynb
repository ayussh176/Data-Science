```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
#this merges on the basis of city
df3 = pd.merge(df1, df2, on="city")
df3
```

```python
df4 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df4
```

```python
df5 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df5
```

```python
df6=pd.merge(df1,df2,on="city",how="inner")
df6
```

```python
df6=pd.merge(df4,df5,on="city",how="outer")
df6
```

```python
df6=pd.merge(df4,df5,on="city",how="outer",indicator=True)
df6
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
#this adds the suffixes to the same type of column but different dataframe
df3= pd.merge(df1,df2,on="city",how="outer", suffixes=('_first','_second'))
df3
```
