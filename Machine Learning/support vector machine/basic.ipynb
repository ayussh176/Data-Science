```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
#svm is a supervised learning model that can be used for classification or regression
#it is more precise than other models like logistic regression
#it is also more complex and requires more computational resources
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df.head()
```

```python
df[df.target == 0]
df.head()
```

```python
iris.target_names
df['flower_name'] = df['target'].apply(lambda x: iris.target_names[x])
df.head()
#this is a simple way to add a new column to the dataframe based on the target values
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
#this creates three dataframes, one for each target class
```

```python
df0.head()
```

```python
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='red', label='setosa',marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='blue', label='versicolor',marker='.')
plt.show()
```

```python
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color='red', label='setosa',marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='blue', label='versicolor',marker='.')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
#split the data into training and testing sets
```

```python
X = df.drop(['target', 'flower_name'], axis=1)
y = df['target']
#this separates the features from the target variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(X_train), len(X_test)
```

```python
from sklearn.svm import SVC
#SVC is the support vector classifier, which is a type of SVM
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict([[5.1,3.5,1.4,0.2]])
```
