```python
import pandas as pd
df = pd.read_excel('survey.xlsx')
df
```

```python
#crosstab hepls to get the frequency of certail value or object
pd.crosstab(df.Nationality,df.Handedness)
```

```python
#if we do margin as true this will help to get total of each row and column
pd.crosstab(df.Sex,df.Handedness,margins=True)
```

```python
pd.crosstab(df.Handedness,df.Nationality,margins=True)
```

```python
import numpy as np
pd.crosstab(df.Sex,df.Handedness,values=df.Age,aggfunc=np.average)
```

```python
#this tell percentage
pd.crosstab(df.Handedness,df.Nationality,normalize='index')
```
