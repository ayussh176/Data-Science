```python
import pandas as pd
df = pd.read_csv('weather_data.csv')
df
```

```python
#the day column is read as string
#to change that we use "phare" 
df = pd.read_csv('weather_data.csv',parse_dates=['day'])# now the day is considered as date
df.set_index('day',inplace=True) # this sets the index
df
```

```python
new_df = df.fillna(0)
new_df # all NaN get replace by 0
```

```python
new_df = df.fillna({'temperature':0,'windspeed':0,'event':'no event'})
new_df #here we have passed a dictionary to specify the values for each columne
```

```python
new_df = df.fillna(method='ffill')#this carry forwarded the previous cell value "bfill" use for copying next cell value
new_df
```

```python
new_df = df.fillna(method='bfill',limit=1)#here limit means copy only 1 time
new_df
```

```python
new_df = df.interpolate()#it give the most likely value of missing place
new_df
#by default it gives linear value
```

```python
new_df = df.interpolate(method='time')
new_df
```

```python
new_df = df.dropna()#drops the row with empty values
new_df
```

```python
new_df = df.dropna(how='all')#gives conditional power to give condition
new_df
```

```python
new_df = df.dropna(thresh=2)#threash means non empty value
new_df
```

```python
dt = pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(dt)
df=df.reindex(idx)
df
```

```python
import numpy as np
rp = pd.read_csv('weather_data_replace.csv')
rp
```

```python
new_rp = rp.replace(-99999,np.nan)#here we have replaced the value -99999 to NaN
new_rp
```

```python
new_rp = rp.replace([-99999],np.nan)#here we have replaced the value -99999 to NaN
new_rp
```

```python
new_rp = rp.replace({'temperature':-99999,'windspeed':-99999,'event':['no event',0]},np.nan)
new_rp
```

```python
# to replace the F and C and mph we use regex which detects the pattern
new_rp = rp.replace({'temperature':'[A-Za-z]','windspeed':'[A-Za-z]'},'',regex=True)
new_rp
#here we have removed the unwanted items
```
